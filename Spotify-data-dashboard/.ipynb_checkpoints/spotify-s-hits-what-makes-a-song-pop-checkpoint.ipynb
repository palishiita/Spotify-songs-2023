## **4. Danceability, emotional Valence, Bpm VS. Impact on song popularity.**

#### **Variables Explanation**:
* **danceability**: Percentage indicating how suitable the song is for dancing
* **valence**: Positivity of the song's musical content
* **bpm**: Beats per minute, a measure of song tempo

#### **Steps**:
1. **Import Necessary Libraries**: `pandas` for data manipulation, `matplotlib` for plotting, and `seaborn` for statistical visualizations.
*Repeat the importing libraries step here so you can use each section snippet code individually.*

2. **Data Cleaning and Pre-processing:**
Handle missing or non-numeric entries to facilitate accurate analysis.

3. **Correlation Computation and Heatmap Visualization:**
Implement visualizations to observe the distributions of 'danceability', 'valence', 'BPM', and 'streams'.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/kaggle/input/top-spotify-songs-2023/spotify-2023.csv'
spotify_data = pd.read_csv(file_path, encoding='ISO-8859-1')

# The stream data of "Love Grows" is corrupt
# Convert columns to numeric type, coercing errors to NaN
spotify_data['bpm'] = pd.to_numeric(spotify_data['bpm'], errors='coerce')
spotify_data['danceability_%'] = pd.to_numeric(spotify_data['danceability_%'], errors='coerce')
spotify_data['valence_%'] = pd.to_numeric(spotify_data['valence_%'], errors='coerce')
spotify_data['streams'] = pd.to_numeric(spotify_data['streams'], errors='coerce')

# Handle NaN values by removing them
spotify_data = spotify_data.dropna(subset=['bpm', 'danceability_%', 'valence_%', 'streams'])

# Correlation matrix
corr_matrix = spotify_data[['bpm', 'danceability_%', 'valence_%', 'streams']].corr()

# Create a custom color palette with potify green
cmap_spotify = sns.light_palette("#1DB954", as_cmap=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, cmap=cmap_spotify, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

# Titles and labels
plt.title('Correlation Heatmap of Song Characteristics and Popularity', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show plot
plt.show()
```

## **5. Analysis of Acousticness, Instrumentalness, Liveness, and Speechiness in Relation to Popularity on Spotify**

#### **Variables Explanation**:
* **acousticness**: Amount of acoustic sound in the song
* **instrumentalness**: Amount of instrumental content in the song
* **liveness**: Presence of live performance elements
* **speechiness**: Amount of spoken words in the song

#### **Steps**:
1. **Import Necessary Libraries**: `pandas` for data manipulation, `matplotlib` for plotting, and `seaborn` for statistical visualizations.

*Repeat the importing libraries step here so you can use each section snippet code individually.*

2. **Load and Preprocess Data**

3. **Calculate Correlations**

4. **Visualize Relationships with Scatter Plots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load and clean data
data = pd.read_csv('/kaggle/input/top-spotify-songs-2023/spotify-2023.csv', encoding='ISO-8859-1')
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
data = data.dropna(subset=['streams'])
data['streams'] = data['streams'].astype(int)

# Calculate the correlation
correlation_columns = ['acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
correlations = data[['streams'] + correlation_columns].corr().loc['streams', correlation_columns]

# Set the Visualization
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))
features = ['acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']

# Create subplots for each feature
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.scatterplot(x=data[feature], y=data['streams'])
    plt.title(f'Streams vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Streams')
    # Displaying the correlation on the plot
    plt.annotate(f"Corr: {correlations[feature]:.2f}", 
                 xy=(0.7, 0.9), xycoords='axes fraction', fontsize=12)

plt.tight_layout()
plt.show()
```

## **6. Exploring Song Popularity Across Various Music Platforms: Part 1**

#### **Approach**:
1. Data Preprocessing:
Handle missing or non-numeric entries to ensure accurate clustering.
Normalize the data to mitigate the impact of the scale of variables.

2. Hierarchical Clustering & Dendrogram:
Utilize hierarchical clustering to build a dendrogram, aiding in visual identification of natural groupings within the data.
Determine the optimal number of clusters by interpreting the dendrogram.

3. PCA for Data Visualization:
Implement Principal Component Analysis (PCA) to reduce the dimensionality of the data, enabling visualization in a 2D space.
Visualize clusters to observe how songs group together based on the chosen features.

4. Cluster Analysis:
Analyze the characteristics of each cluster by examining the mean values of the features.

#### **Features Utilized**:
in_spotify_playlists
in_spotify_charts
in_apple_playlists
in_apple_charts
in_deezer_playlists
in_deezer_charts
in_shazam_charts
streams

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#Load and clean data
data = pd.read_csv('/kaggle/input/top-spotify-songs-2023/spotify-2023.csv', encoding='ISO-8859-1')
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
data = data.dropna(subset=['streams'])
data['streams'] = data['streams'].astype(int)

# Features to use for clustering
features = ['acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%', 'danceability_%', 'valence_%', 'bpm']

# Scaling the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[features])

sse = []
possible_k_values = range(2, 11)

for k in possible_k_values:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(data_scaled)
    sse.append(kmeans.inertia_)

# Plotting the Elbow Method graph
plt.figure(figsize=(10, 6))
sns.lineplot(x=possible_k_values, y=sse, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('k')
plt.ylabel('Sum of Squared Distances')
plt.show()

# Apply K-Means with the optimal k
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=42)
data['cluster'] = kmeans.fit_predict(data_scaled)

# Analyzing the clusters
cluster_summary = data.groupby('cluster')[features].mean()

# Visualizing the clusters
# For example, let's plot 'danceability' vs 'valence', colored by cluster
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['danceability_%'], y=data['valence_%'], hue=data['cluster'], palette='viridis', s=50)
plt.title('Clusters Visualized by Danceability and Valence')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load and clean data
data = pd.read_csv('/kaggle/input/top-spotify-songs-2023/spotify-2023.csv', encoding='ISO-8859-1')
data['streams'] = pd.to_numeric(data['streams'], errors='coerce')
data = data.dropna(subset=['streams'])
data['streams'] = data['streams'].astype(int)

# Features for clustering
features = [
    'in_spotify_playlists', 'in_spotify_charts',
    'in_apple_playlists', 'in_apple_charts',
    'in_deezer_playlists', 'in_deezer_charts',
    'in_shazam_charts', 'streams'
]

# Convert string columns to integer
data[features] = data[features].apply(lambda x: pd.to_numeric(x.astype(str)
                                                .str.replace(',',''), errors='coerce'))

# Handling NaN values after conversion
data = data.dropna(subset=features)

# Scaling the features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[features])

# Creating a Dendrogram
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(data_scaled, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Songs')
plt.ylabel('Euclidean Distances')
plt.show()

# Applying Hierarchical Clustering
# (Choose n_clusters based on the dendrogram)
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
data['hierarchical_cluster'] = hc.fit_predict(data_scaled)

# Analyzing the Clusters
cluster_summary = data.groupby('hierarchical_cluster')[features].mean()
print(cluster_summary)

# Apply PCA and fit the features
pca = PCA(n_components=2)
principal_components = pca.fit_transform(data_scaled)

# Create a DataFrame with the principal components
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Add the cluster label obtained from hierarchical clustering to this DataFrame
pc_df['Cluster'] = data['hierarchical_cluster']

# Visualizing the 2D projection
plt.figure(figsize=(10, 7))
sns.scatterplot(x=pc_df['PC1'], y=pc_df['PC2'], hue=pc_df['Cluster'], palette='viridis', s=50)
plt.title('2D PCA of Song Clusters')
plt.show()
```

**Cluster 0** has a medium number of Spotify playlists, the highest rank in Spotify charts, medium-level presence in Apple and Deezer playlists, and charts, and a high rank in Shazam charts. The streams are also quite significant, making this a cluster of fairly popular songs.

**Cluster 1** has the lowest presence and ranks across all platforms and the lowest number of streams, implying these are the less popular songs.

**Cluster 2** has the highest presence in Spotify, Apple, and Deezer playlists, a medium rank in their respective charts, and a decent rank in Shazam charts. This cluster also has the highest number of streams, making it the cluster of the most popular songs.

## **6. Exploring Song Popularity Across Various Music Platforms: Part 2**

#### **Background**:
Through clustering, we grouped songs based on certain features (e.g., streaming numbers, presence in various platform charts, and playlists) to identify patterns or similarities among them. Each cluster represents a group of songs that share similarities across these features, potentially indicating similar user behavior and popularity patterns across different platforms.

#### **What is a Cluster in This Context?**
In the realm of our analysis, a **cluster** is a collection of songs that share similar characteristics related to their popularity and presence on various music platforms, such as Spotify and Apple Music. Songs within the same cluster exhibit similar behaviors in terms of streams, presence in charts, and inclusion in playlists across these platforms.

#### **Exploring Top Songs and Artists:**
We delve into each cluster to identify the top songs and artists, aiming to glean insights into the specific tracks and musicians that characterize each group. This exploration may reveal specific artists or tracks that tend to align with the popularity and platform presence patterns encapsulated by each cluster.

#### **Steps**:
1. **Identify Top Songs and Artists**:
   Extract and list the most streamed songs and most recurrent artists within each cluster to understand the kind of music that is popular within the respective group.
   
2. **Analyze & Interpret**:
   Investigate the listed songs and artists, seeking patterns or commonalities. For instance, do certain artists tend to appear in a specific cluster? Are there observable trends in the genre or style of the top songs?

#### **Significance**:
Understanding the top songs and artists within each cluster not only unveils the specific tracks and musicians that define each group but also potentially highlights trends in user behavior and music popularity across different platforms.

```python
# Number of top songs and artists to display
top_n = 10

# Loop through each cluster
for cluster in data['hierarchical_cluster'].unique():
    subset_data = data[data['hierarchical_cluster'] == cluster]
    
    print(f"\nCluster {cluster}:")
    print(f"Top {top_n} Songs:")
    print(subset_data.nlargest(top_n, 'streams')['track_name'])
    
    print(f"\nTop {top_n} Artists:")
    print(subset_data['artist(s)_name'].value_counts().head(top_n))
```
